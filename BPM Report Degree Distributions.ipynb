```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import powerlaw
```

```python
data_path = r'/Users/Owner/Documents/Work_transfer/Data/GCconnex/'
```

```python
#Importing the Degree value from Gephi, then dropping the groups for BPM
df = pd.read_csv(data_path+'BPM Degree Distribution.csv')
df = df[df.department != 'group']
df = df.reset_index()
df = df.drop('index', 1)
df['odfreq'] = df['Out-Degree'].value_counts()
```

```python
plt.scatter(np.arange(len(df)), df['odfreq'])
plt.show()
```

```python
results = powerlaw.Fit(df['odfreq'], discrete=True)
```

```python
#Same thing, but with GC2.0 Tools
gc = pd.read_csv(data_path+'GC2.0 DD.csv')
gc = gc[gc.department != 'group']
gc = gc.reset_index()
gc = gc.drop('index', 1)
gc['odfreq'] = gc['Out-Degree'].value_counts()
```

```python
plt.plot(gc['odfreq'])
plt.show()
```
