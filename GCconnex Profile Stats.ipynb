```python
#This is the beginnings of making a script that can take a general dataset
#Containing data on whether an individual in a specific department has filled out the 'About me' Section
# And/Or has created an avatar
import numpy as np
import pandas as pd
import csv
```

```python
data_path = r"/Users/Owner/Documents/Work_transfer/Data/GCconnex/"
```

```python
df1 = pd.read_csv(data_path+'gcconnex_user_about_icon.csv')
```

```python
#Here we will have to insert the code to group people by a specific column
#That contains the department that they're in
#The current data just contains everybody
```

```python
#Renaming the About me Icon column, remembering that 55 = About me and 73 = Prof Pic
#This part will have to be altered to account for the department column when the right dataset it in
df1.columns = ['Indicator', 'Content Icon', 'Owner GUID', 'Time Created', 'Last Updated']
df1 = df1.drop('Content Icon', 1)
df1 = df1.drop('Time Created', 1)
df1 = df1.drop('Last Updated', 1)
df1 = df1[['Owner GUID', 'Indicator']]
```

```python
#Splitting the Dataset into two Dataframes and removing duplicate entries
#If somebody edited their about me profile multiple times they will appear multiple times
df2 = df1.loc[df1['Indicator'] == 73]
df1 = df1.loc[df1['Indicator'] == 55]
df2.drop_duplicates(inplace=True)
df1.drop_duplicates(inplace=True)
```

```python
#Merging the two dataframes into one single dataframe containing telling whether that individual
#Has filled uploaded an avatar or filled out the "about me" section
df = pd.merge(df1, df2, on='Owner GUID', how='outer')
df.columns = ['Owner GUID', 'About me', 'Avatar']
```

```python
df.set_index(df['Owner GUID'], inplace=True)
df = df.drop('Owner GUID', 1)
```

```python
df['Both'] = df['About me'] + df['Avatar']
```

```python
def f(x):
    if x == 128:
        return 'Y'
    else:
        return 'N'

def g(x):
    if x == 55:
        return 'Y'
    else:
        return 'N'
    
def h(x):
    if x == 73:
        return 'Y'
    else:
        return 'N'

df['Both'] = df['Both'].apply(f)
df['About me'] = df['About me'].apply(g)
df['Avatar'] = df['Avatar'].apply(h)
```
